```python
import pandas as pd
df = pd.read_csv('bmw.csv')
df.head()
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
plt.scatter(df['Milage'], df['Price'])
```

```python
plt.scatter(df['Age(in Year)'], df['Price'])
```

```python
x = df[['Milage', 'Age(in Year)']]
y = df['Price']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
clf = LinearRegression() #this use to train the model
```

```python
clf.fit(x_train, y_train) #Fit the data into linear regression model 
clf.predict(x_test)       # Predict price of car
```

```python
clf.score(x_test, y_test)
```
