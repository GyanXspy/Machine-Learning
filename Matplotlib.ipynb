```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

```python
#Ploting the data 
#sin wave 
plt.plot(x, y)
plt.show()
```

```python
#Cos wave 
plt.plot(x,z)
plt.show()
```

```python
#Adding title x-axis & y-axis labels
plt.plot(x, y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#parabola
x = np.linspace(-10, 10, 20)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r+')
```

```python
plt.plot(x, y, 'g.')
```

```python
plt.plot(x, y, 'yx')
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g>')
```

```python
plt.plot(x, np.cos(x), 'r>')
```

```python
plt.plot(x, np.sin(x), 'g.')
plt.plot(x, np.cos(x), 'r*')
```

```python
#Bar
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
language = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 70, 30, 70]
ax.bar(language, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
#Pie chart
fig1 = plt.figure()
ax = fig1.add_axes([0, 0, 1, 1])
language = ['English', 'French', 'Spanish', 'Latin', 'German']
people = [100, 50, 70, 30, 70]
ax.pie(people, labels = language, autopct = '%1.1f%%')
plt.show()
```

```python
#Scatter plot
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x, y, color = 'r')
ax.scatter(x, z, color = 'g')
```

```python
fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x, y, z, c = z, cmap = 'Blues')
```
